# Generate Dataset Contains Space, English, Greek, Symbols

## Combination of Greek Words

```python
import pandas as pd
import csv
df = pd.read_csv('Greek_corpus.txt', sep="\t", header=None, quoting=csv.QUOTE_NONE)
```

```python
import pandas as pd
import csv
df_dict = pd.read_csv('greek_eng_dict.txt', sep="\t", header=None, quoting=csv.QUOTE_NONE)
dictionary = pd.read_csv('greek_eng_dict.txt', sep="\t", header=None, quoting=csv.QUOTE_NONE)
df_dict = df_dict.iloc[110:]
df_dict = df_dict.reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df_dict
```

```python
import random

with open('augment.txt', 'w') as z:

    for i in range(100):

        # Pick Words in Corpus

        f = random.randint(0,len(df)-1)
        s = random.randint(0,len(df)-1)

        # Pick Symbols in List

        fs = random.randint(0,len(df_dict)-1)
        ss = random.randint(0,len(df_dict)-1)

        # Add Space at Random

        space = random.randint(0,1)

        # Merge
        if space == 1:
            string = df.iloc[f][0] + df_dict.iloc[fs][0] + ' ' + df.iloc[s][0] + df_dict.iloc[ss][0]
        else:
            string = df.iloc[f][0] + df_dict.iloc[fs][0] + df.iloc[s][0] + df_dict.iloc[ss][0]

        if len(string) >= 25:
            string = string[0:25]

        print(string)
    
        z.write(string)
        z.write('\n')
```

## Data Cleansing for English Dataset

```python
import pandas as pd
import csv
df_eng = pd.read_csv('eng.wordlist', sep="\t", header=None, quoting=csv.QUOTE_NONE)
```

```python
df_eng
```

```python
dict_list = list(dictionary[0])

remove_list = []

for i, txt in enumerate(df_eng[0]):
    try:
        for j in txt:
            if j in dict_list:
                pass
            else:
                remove_list.append(i)
                continue
    except:
        remove_list.append(i)
```

```python
df_eng_avalible = df_eng.drop(index=remove_list)
df_eng_avalible = df_eng_avalible.reset_index(drop=True)
```

```python
df_eng_avalible
```

```python
with open('English_corpus.txt', 'w') as z:

    for i in df_eng_avalible[0]:
        z.write(i)
        z.write('\n')
```
